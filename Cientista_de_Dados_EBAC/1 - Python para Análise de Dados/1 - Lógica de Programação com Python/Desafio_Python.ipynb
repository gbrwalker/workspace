```python
input_1 = input("Digite o primeiro valor:")
input_2 = input("Digite o segundo valor:")
num1 = int(input_1)
num2 = int(input_2)
print("Escolha qual operação deseja realizar, as opções são as seguintes:")
print("1 - Soma (+)")
print("2 - Subtração (-)")
print("3 - Divisão (/)")
print("4 - Multiplicação (*)")
print("5 - Resto da Divisão (%)")
simbolo = input("Digite o número referente a operação que deseja realizar:")
operacao = int(simbolo)
if operacao == 1:
  resultado = num1 + num2
  print(f"A Soma entre os 2 numeros é : {resultado}")
elif operacao == 2:
  resultado = num1 - num2
  print(f"A Subtração entre os 2 numeros é : {resultado}")
elif operacao == 3:
  resultado = num1 / num2
  print(f"A Divisão entre os 2 numeros é : {resultado}")
elif operacao == 4:
  resultado = num1 * num2
  print(f"A Multiplicação entre os 2 numeros é : {resultado}")
elif operacao == 5:
  resultado = num1 % num2
  print(f"O Resto da divisão entre os 2 numeros é : {resultado}")
else:
  print("Você escolheu uma opção invalida, comece novamente!")
```
